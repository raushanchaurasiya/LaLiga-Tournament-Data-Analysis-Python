### Q1.Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
#replacing dash(-) with 0.
laliga_df = laliga_df.replace(to_replace="-",value=0)
```

```python
laliga_df.info()
```

**Observations:**
* We need to change the datatypes of various columns as we can see that the datatype of columns are showing object but we have numerical data.
* So, we will first change the datatype so that its easy for us to perform operations.

```python
#changing datatype from object to int
laliga_df[['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst','Champion','Runner-up','Third','Fourth','Fifth','Sixth','T']] = laliga_df[['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst','Champion','Runner-up','Third','Fourth','Fifth','Sixth','T']].apply(pd.to_numeric)
print(laliga_df.dtypes)
```

### Q4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 

```python
def goal_diff_count(laliga_df):
    laliga_df['goal_diff'] = laliga_df['GoalsFor'] - laliga_df['GoalsAgainst']
    print(laliga_df[['Team','goal_diff']])
    print("******************************************\n")
    
    print("Team with maximum Goal difference")
    print(laliga_df[laliga_df['goal_diff'] == laliga_df['goal_diff'].max()])
    
    print("******************************************\n")
    print("Team with minimum Goal difference")
    print(laliga_df[laliga_df['goal_diff'] == laliga_df['goal_diff'].min()])

goal_diff_count(laliga_df)
```

**Observations :**
* **Real Madrid Team** has a Maximum Goal difference of 2807.It means it has done 2807 more than the other teams has done goal to them.
* **Racing Santander** Team has minimum Goal difference of -525. It means other team has done 525 more goals to them than total of goals done by Racing Santander.

### Q5. Create a new column with name “Winning Percent” and append it to the data set .
